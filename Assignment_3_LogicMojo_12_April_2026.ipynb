Q1. Understanding Feature Distributions (Iris Dataset)

Context: Before training any ML model, we analyze feature distributions.

Dataset: Load Iris dataset from sklearn

Tasks:

1. Extract feature matrix as NumPy array
2. Compute:
Mean
Median
Standard deviation
Variance (for each feature)
3. Identify:
Which feature has highest variability and why it matters in ML
4. Convert any one feature into shape (n,1) and explain why ML models expect this format

```python
import pandas as pd
import numpy as np 
from sklearn.datasets import load_iris

#1. Extract feature matrix as NumPy array
iris_dataset = load_iris()
print ("Iris Feature matrix as NumPy array (displaying first 10 rows):")
print (iris_dataset.data[:10])
#2. Compute:
#Mean
#Median
#Standard deviation
#Variance (for each feature)
mean = np.mean(iris_dataset.data, axis=0)
median = np.median(iris_dataset.data, axis=0)
std_dev = np.std(iris_dataset.data, axis=0)
variance = np.var(iris_dataset.data, axis=0)
print("\nMean of each feature:")
print(mean)
print("\nMedian of each feature:")
print(median)
print("\nStandard Deviation of each feature:")
print(std_dev)
print("\nVariance of each feature:")
print(variance)
#3. Identify:Which feature has highest variability and why it matters in ML
high_var_feature_index = np.argmax(variance)
iris_feature = iris_dataset.feature_names
high_var_feature = iris_feature[high_var_feature_index]
print(f"\nFeature with the highest variability is {high_var_feature} with Variance: {variance[high_var_feature_index]:.4f}.")
print("\nWhy it matters in ML?")
print('''Features that change more across data points can help the model identify patterns and differences.
If the variation is related to the target, it improves prediction.
Features with very low variability usually add little value and can be ignored.''')
```

Q3. Feature Scaling using NumPy (Very Important)

Context: Most ML models require normalized data.

Dataset: Iris (NumPy array)

Tasks:

1. Apply standardization:

$$
X' = \frac{X - \mu}{\sigma}
$$
	​2. 

Verify:
Mean ≈ 0
Std ≈
on

```python
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
X = iris.data

# 1. Apply standardization: X' = (X - μ) / σ
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)
X_standardized = (X - mean) / std

print("Original data (first 5 rows):")
print(X[:5])
print("\nStandardized data (first 5 rows):")
print(X_standardized[:5])

# 2. Verify: Mean ≈ 0, Std ≈ 1
mean_standardized = np.mean(X_standardized, axis=0)
std_standardized = np.std(X_standardized, axis=0)

np.set_printoptions(precision=6, suppress=True)

print("\nVerification after standardization:")
print("Mean of each feature:", mean_standardized)
print("Std of each feature:", std_standardized)
```

Q6. Handling Missing Data (Critical ML Step)

Context: Models cannot handle null values.

Dataset: Titanic

Tasks:

1. Fill missing age using median
2. Fill missing embarked using mode
3. Drop deck column
4. Explain:
Why different strategies are used for different columns

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

# First check what's missing
print("Missing values before cleaning:")
print(titanic.isnull().sum())

# 1. Fill missing age with median
age_median = titanic['age'].median()
print(f"\nMedian age: {age_median}")
titanic['age'] = titanic['age'].fillna(age_median)

# 2. Fill missing embarked with mode (most common value)
embarked_mode = titanic['embarked'].mode()[0]
print(f"Most common embark port: {embarked_mode}")
titanic['embarked'] = titanic['embarked'].fillna(embarked_mode)

# 3. Drop the deck column (has too many missing values)
titanic.drop('deck', axis=1, inplace=True)

print("\nMissing values after cleaning:")
print(titanic.isnull().sum())

# 4. Why different strategies?
print("\n" + "="*50)
print("Why use different strategies?")
print("\nAge: Used median because it's a number")
print("- We can't use a category for age")
print("- Median is better than mean (less affected by outliers)")
print("\nEmbarked: Used mode because it's a category")
print("- Can't use a number for a port name")
print("- Mode picks the most common port")
print("\nDeck: Just dropped it")
print("- Too many missing values (too incomplete)")
print("- Not worth trying to fix")
```

Q7. Feature Engineering for Model Improvement

Context: Raw data is rarely useful—features must be created.

Dataset: Titanic

Tasks:

1. Create:
family_size = sibsp + parch + 1
2. Create:
is_alone (binary feature)
3. Compute survival rate by is_alone
4. Explain:
Why engineered features improve ML models

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 1. Create family_size
titanic['family_size'] = titanic['sibsp'] + titanic['parch'] + 1 #themselves
print("Created family_size:")
print(titanic[['sibsp', 'parch', 'family_size']].head(10))

# 2. Create is_alone (1 if family_size is 1, else 0)
titanic['is_alone'] = (titanic['family_size'] == 1).astype(int)
print("\nCreated is_alone:")
print(titanic[['family_size', 'is_alone']].head(10))

# 3. Survival rate by is_alone
print("\n" + "="*50)
print("Survival rates:")
survival_by_alone = titanic.groupby('is_alone')['survived'].mean()
print(survival_by_alone)
print(f"\nAlone passengers survived: {survival_by_alone[1]:.2%}")
print(f"Not alone passengers survived: {survival_by_alone[0]:.2%}")

# 4. Why does this help?
print("\n" + "="*50)
print("Why these features help:")
print("- family_size tells the model how many people traveled together")
print("- is_alone is simpler - just 0 or 1")
print("- The model can learn that people traveling alone had different survival rates")
print("- These are easier for the model to use than sibsp and parch separately")
print("- Sometimes combining info creates better patterns for the model to learn")
```

Q8. Group-Based Insights (Customer Segmentation Logic)

Context: ML models often depend on aggregated patterns.

Dataset: Tips

Tasks:

1. Compute:
Average tip by day
Total bill by gender
2. Create new feature:
tip_percentage
3. Identify:
Which day has highest tipping behavior
4. Explain:
How this helps in segmentation models

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')

# 1. Average tip by day
print("Average tip by day:")
avg_tip_by_day = tips.groupby('day', observed=True)['tip'].mean()
print(avg_tip_by_day)

# Total bill by gender
print("\nTotal bill by gender:")
total_bill_by_gender = tips.groupby('sex', observed=True)['total_bill'].sum()
print(total_bill_by_gender)

# 2. Create tip_percentage
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100
print("\nTip percentage (first 5 rows):")
print(tips[['total_bill', 'tip', 'tip_percentage']].head())

# 3. Which day has highest tipping
print("\n" + "="*50)
avg_tip_pct_by_day = tips.groupby('day', observed=True)['tip_percentage'].mean()
print("Average tip percentage by day:")
print(avg_tip_pct_by_day)
highest_tipping_day = avg_tip_pct_by_day.idxmax()
print(f"\nHighest tipping day: {highest_tipping_day} ({avg_tip_pct_by_day[highest_tipping_day]:.1f}%)")

# 4. How does this help segmentation?
print("\n" + "="*50)
print("How this helps in segmentation:")
print("- We can split customers into groups (by day, gender, etc)")
print("- Saturdays and Sundays might be different from weekdays")
print("- Males and females might tip differently")
print("- If we know these patterns, we can predict who tips more")
print("- The model can learn: 'customers on Saturday tip better'")
print("- We can target different strategies for different segments")
```

Q9. Advanced Filtering + Subsetting

Context: Training data is often filtered based on business rules.

Dataset: Tips

Tasks:

1. Filter:
total_bill > 20 AND tip < 3
2. Select only relevant columns
3. Analyze:
Is this segment under-tipping?
4. Explain:
How such filtering helps anomaly detection

```python
import pandas as pd
import seaborn as sns

tips = sns.load_dataset('tips')

# 1. Filter: total_bill > 20 AND tip < 3
filtered = tips[(tips['total_bill'] > 20) & (tips['tip'] < 3)]
print("Filtered data (high bill, low tip):")
print(filtered)

# 2. Select only relevant columns
relevant = filtered[['total_bill', 'tip', 'day', 'sex', 'size']]
print("\nRelevant columns only:")
print(relevant)

# 3. Analyze - is this segment under-tipping?
print("\n" + "="*50)
print("Analysis:")
print(f"Count: {len(filtered)} cases")
print(f"Average bill: ${filtered['total_bill'].mean():.2f}")
print(f"Average tip: ${filtered['tip'].mean():.2f}")

overall_avg_tip_pct = (tips['tip'].sum() / tips['total_bill'].sum()) * 100
filtered_tip_pct = (filtered['tip'].sum() / filtered['total_bill'].sum()) * 100
print(f"\nOverall tip percentage: {overall_avg_tip_pct:.1f}%")
print(f"This segment's tip percentage: {filtered_tip_pct:.1f}%")
print("\nYes, they're definitely under-tipping!")

# 4. How does this help anomaly detection?
print("\n" + "="*50)
print("How filtering helps detect anomalies:")
print("- We found a small group with weird behavior (high bill + low tip)")
print("- This is suspicious - might be complaints or errors")
print("- We can flag these for review")
print("- The model learns: this pattern is different from normal")
print("- Real-world use: find fraud, mistakes, or unhappy customers")
print("- Filter first, then investigate the outliers")
```

Q10. Data Transformation for Model Input (NumPy)

Context: ML models require structured numeric arrays.

Tasks:

1. Create two arrays of shape (3,2)
2. Perform:
Vertical stacking
Horizontal stacking
3. Reshape into (2,6)
4. Explain:
Why reshaping is required in ML pipelines

```python
import numpy as np

# 1. Create two arrays of shape (3,2)
arr1 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])

arr2 = np.array([[7, 8],
                 [9, 10],
                 [11, 12]])

print("Array 1:")
print(arr1)
print("\nArray 2:")
print(arr2)

# 2. Vertical stacking (stack on top of each other)
v_stacked = np.vstack([arr1, arr2])
print("\n" + "="*50)
print("Vertical stacking (vstack):")
print(v_stacked)
print(f"Shape: {v_stacked.shape}")

# Horizontal stacking (stack side by side)
h_stacked = np.hstack([arr1, arr2])
print("\nHorizontal stacking (hstack):")
print(h_stacked)
print(f"Shape: {h_stacked.shape}")

# 3. Reshape into (2,6)
print("\n" + "="*50)
reshaped = v_stacked.reshape(2, 6)
print("Reshaped to (2,6):")
print(reshaped)
print(f"Shape: {reshaped.shape}")

# 4. Why reshaping matters
print("\n" + "="*50)
print("Why reshaping is important:")
print("- Models expect data in specific shapes")
print("- If you have 12 values but need (2,6), you reshape")
print("- Changing from (6,2) to (2,6) rearranges the same data")
print("- It's like reorganizing rows and columns")
print("- Most models need data as (samples, features)")
print("- If your data is the wrong shape, the model won't work")
```

Q11. End-to-End ML Preprocessing Pipeline (Titanic)

Context: Prepare dataset for ML model training

Tasks:

1. Load dataset
2. Handle missing values
3. Encode categorical variables (sex, embarked)
4. Normalize numerical features
5. Convert to NumPy array

Output:

Final feature matrix ready for model training

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# 1. Load dataset
titanic = sns.load_dataset('titanic')
print("Dataset loaded:", titanic.shape)

# 2. Handle missing values
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['embarked'] = titanic['embarked'].fillna(titanic['embarked'].mode()[0])
titanic.drop('deck', axis=1, inplace=True)
print("Missing values handled")

# 3. Encode categorical variables
# Sex: male/female to 1/0
titanic['sex'] = (titanic['sex'] == 'male').astype(int)

# Embarked: C/Q/S to 0/1/2
embarked_map = {'C': 0, 'Q': 1, 'S': 2}
titanic['embarked'] = titanic['embarked'].map(embarked_map)
print("Categorical variables encoded")

# 4. Select features and normalize
features = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'embarked']
X = titanic[features]

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
print("Features normalized")

# 5. Convert to NumPy array and show result
print(f"\nFinal feature matrix shape: {X_normalized.shape}")
print("First 5 rows:")
print(X_normalized[:5])
```

Q12. Feature Selection using Statistical Understanding

Context: Not all features improve model performance.

Dataset: Iris

Tasks:

1. Compute correlation matrix
2. Identify highly correlated features
3. Drop redundant features
4. Explain:
Impact on overfitting and model performance

```python
import pandas as pd
import seaborn as sns

# Load iris data
tips = sns.load_dataset('iris')

# 1. Compute correlation matrix using only numeric columns
corr_matrix = tips.corr(numeric_only=True)
print("Correlation matrix:")
print(corr_matrix)

# 2. Identify highly correlated features
# We consider absolute correlation > 0.8 as high.
high_corr = corr_matrix.abs() > 0.8
print("\nHighly correlated pairs (abs value > 0.8):")
print(high_corr)

# 3. Drop redundant features
# Petal length and petal width are usually very correlated with sepal measurements.
# Let's keep the two clearest features and drop one redundant one.
selected_features = tips[['sepal_length', 'sepal_width', 'petal_length']]
print("\nSelected features after dropping a redundant one:")
print(selected_features.head())

# 4. Explain impact on overfitting and model performance
print("\nWhy drop redundant features?")
print("- Redundant features add noise without giving new information.")
print("- Fewer features reduce the chance of overfitting.")
print("- Models train faster with fewer columns.")
print("- If two columns tell the same story, keeping both does not help much.")
print("- Good feature selection makes the model simpler and more stable.")
```

Q13. Custom Feature Transformation using apply()

Context: Real ML requires custom logic transformation.

Dataset: Titanic

Tasks:

1. Create age_group:
Child / Adult / Senior
2. Use .apply()
3. Compute survival rate per group
4. Interpret:
Which segment has highest survival likelihood

```python
import pandas as pd
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# 1. Create age_group using apply()
def age_group(age):
    if age < 18:
        return 'Child'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

# Fill missing age first so apply() works without errors
titanic['age'] = titanic['age'].fillna(titanic['age'].median())
titanic['age_group'] = titanic['age'].apply(age_group)

# 2. Show the new column
def print_group_sample():
    print(titanic[['age', 'age_group']].head(10))

print_group_sample()

# 3. Compute survival rate per group
survival_by_age_group = titanic.groupby('age_group')['survived'].mean()
print('\nSurvival rate by age group:')
print(survival_by_age_group)

# 4. Interpret the result
best_group = survival_by_age_group.idxmax()
print(f"\nHighest survival likelihood: {best_group}")
print('\nWhat this means:')
print('- We grouped passengers into Child, Adult, Senior')
print('- Then we checked who survived most often in each group')
print('- The group with the highest rate is the safest segment')
print('- This helps the model know which age category matters most')
```

Q14. Error Analysis Simulation (Model Evaluation Thinking)

Context: Understanding prediction errors is key in ML.

Tasks:

1. Generate 1000 random error values using NumPy
2. Compute:
Mean
Standard deviation
3. Identify outliers:
Using mean ± 2*std
4. Explain:
What high variance means in model performance

```python
import numpy as np

# 1. Generate 1000 random error values using NumPy
errors = np.random.normal(loc=0, scale=1, size=1000)

# 2. Compute mean and standard deviation
mean_error = np.mean(errors)
std_error = np.std(errors)
print(f"Mean error: {mean_error:.4f}")
print(f"Standard deviation: {std_error:.4f}")

# 3. Identify outliers using mean ± 2*std
lower_bound = mean_error - 2 * std_error
upper_bound = mean_error + 2 * std_error
outliers = errors[(errors < lower_bound) | (errors > upper_bound)]

print(f"\nOutliers count: {len(outliers)}")
print("Example outliers:")
print(outliers[:10])

# 4. Explain what high variance means in model performance
print("\nWhy high variance matters:")
print("- High variance means errors are spread out a lot.")
print("- The model is not consistent; sometimes it makes small errors, sometimes large ones.")
print("- This is bad because it means the model is unreliable on new data.")
print("- In ML, lower variance usually means more stable predictions.")
```

Q15. Business + ML Insight Problem (Tips Dataset)

Context: Convert data analysis into decision-making

Tasks:

1. Create:
tip_percentage
2. Group by:
day and time
3. Identify:
Highest revenue segment
4. Provide:
Recommendation for business strategy
How ML model can use this insight

```python
import pandas as pd
import seaborn as sns

# Load tips dataset
tips = sns.load_dataset('tips')

# 1. Create tip_percentage
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

# 2. Group by day and time
summary = tips.groupby(['day', 'time'], observed=True).agg(
    total_revenue=('total_bill', 'sum'),
    avg_tip_percentage=('tip_percentage', 'mean'),
    count=('tip', 'size')
).reset_index()

print('Revenue summary by day and time:')
print(summary)

# 3. Identify highest revenue segment
best_segment = summary.loc[summary['total_revenue'].idxmax()]
print('\nHighest revenue segment:')
print(best_segment)

# 4. Recommendation and ML insight
print('\nRecommendation:')
print(f"- Focus on {best_segment['day']} {best_segment['time']} because it brings the highest bill revenue.")
print('- Use this segment for staffing and menu planning.')
print('- Offer targeted promotions or upsell items during the busiest times.')

print('\nHow ML can use this insight:')
print('- A model can learn day + time patterns to predict revenue. ')
print('- It can also use tip percentage to spot which slots give better customer value.')
print('- This helps the business plan staffing and menu offers more smartly.')
print('- In short, the model can turn the segment into a decision rule.')
```

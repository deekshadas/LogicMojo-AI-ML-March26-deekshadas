## 1. Customer Segmentation Engine

You are given a list of customer dictionaries:

```python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
```

Task:

1. Write a function segment_customer(customer):
2. If spend divided by income is greater than 0.6 return "High Value"
3. If spend divided by income is between 0.3 and 0.6 return "Medium Value"
4. Otherwise return "Low Value"
5. Create a final dictionary in the format {customer_id: segment}

````python
customers = [
    {"id": 1, "age": 25, "income": 50000, "spend": 20000},
    {"id": 2, "age": 45, "income": 120000, "spend": 80000},
    {"id": 3, "age": 35, "income": 70000, "spend": 30000}
]
def segment_customer(customers):
    for customer in customers:
        if customer["spend"] / customer["income"] > 0.6:
            customer["segment"] = "High Value"
        elif customer["spend"] / customer["income"] > 0.3:
            customer["segment"] = "Medium Value"
        else:
            customer["segment"] = "Low Value"
    return {customer["id"]: customer["segment"] for customer in customers}

print(segment_customer(customers))
````

## 2. Text Cleaning Pipeline

You are given:

```python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]
```

Task:

1. Write a function clean_text(text):
2. Remove special characters
3. Remove extra spaces
4. Convert text to lowercase
5. Apply the function to all elements using loop or map
6. Return cleaned list

---

````python
texts = [
    "Hello!!! How are you???",
    "This is    AI     class.",
    "Python@@@is###great"
]

def clean_text(text):
    cleaned_texts=[]
    for line  in text:
        cleaned = re.sub(r'[^a-zA-Z0-9\s]', '', line)
        cleaned = re.sub(r'\s+', ' ', cleaned).strip()
        cleaned_texts.append(cleaned.lower())
    return cleaned_texts


clean_text(texts)
````

## 3. Fraud Detection Rule Engine

You are given:

```python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]
```

Task:

1. Write a function is_fraud(transaction)
2. If amount is greater than 100000 or location is "Unknown" mark as fraud
3. Otherwise mark as safe
4. Use lambda with filter or map
5. Return only fraud transactions


````python
transactions = [
    {"id": 101, "amount": 5000, "location": "India"},
    {"id": 102, "amount": 200000, "location": "Unknown"},
    {"id": 103, "amount": 1000, "location": "India"}
]

def is_fraud(transaction):
    if transaction["amount"] > 100000 or transaction["location"] == "Unknown":
        return True
    return False

fraud_transactions = list(filter(lambda transaction: is_fraud(transaction), transactions))
print("Fraudulent transactions:", fraud_transactions)
````

## 4. Feature Engineering

You are given:

```python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]
```

Task:

1. Write a function feature_engineer(record)
2. Add a new key salary_per_exp
3. salary_per_exp is salary divided by experience
4. Handle division by zero using conditional logic
5. Return updated dataset

---

````python
data = [
    {"salary": 50000, "experience": 2},
    {"salary": 100000, "experience": 5}
]

def feature_engineer(data):
    for record in data:
        if record["experience"] > 0:
            record["salary_per_experience"] = record["salary"] / record["experience"]
        else:
            record["salary_per_experience"] =  None
    return data

print(f"Updated data: {feature_engineer(data)}")
````

## 6. Sequence Pattern Detection

You are given:

```python
data = [10, 20, 30, 25, 40, 50]
```

Task:

1. Write a function detect_increasing_sequence(data)
2. Find the longest continuous increasing sequence
3. Return that sequence as a list

Example output:

```python
[10, 20, 30]
```


````python
data = [10, 20, 30, 25, 40, 50]

def detect_increasing_sequence(data):
    if not data:
        return []

    longest = []
    current = [data[0]]

    for i in range(1, len(data)):
        if data[i] > data[i - 1]:
            current.append(data[i])
        else:
            if len(current) > len(longest):
                longest = current
            current = [data[i]]

    if len(current) > len(longest):
        longest = current

    return longest

print(detect_increasing_sequence(data))
    
````

## 7. Dictionary-Based Recommendation System

You are given:

```python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}
```

Task:

1. Write a function recommend(user)
2. Recommend items based on common items between users
3. Do not recommend items already purchased by the user
4. Use sets and dictionary iteration

---

````python
user_purchases = {
    "user1": ["apple", "banana"],
    "user2": ["banana", "orange"],
    "user3": ["apple", "orange"]
}

def recommend(purchase_history):
    recommendations = {}
    common_items = set(purchase_history['user1']) | set(purchase_history['user2']) | set(purchase_history['user3'])
    for user, items in purchase_history.items():
        recommendations[user] = common_items - set(items)
    return recommendations
    
print("Recommended items for all users:", recommend(user_purchases))
````

## 8. Recursive Decision Tree Simulation

Task:

1. Write a recursive function decision_tree(score)
2. If score is greater than 80 return "Approved"
3. If score is between 50 and 80 call the function again with score plus 10
4. If score is less than 50 return "Rejected"

---


````python
def decision_tree(score):
    #stop condition
    if score <50:
        return "Rejected"
    elif score >80:
        return "Approved"
    else:
        score += 10
        #print(score)
    return decision_tree(score)
    
print("Decision tree result for score 45:", decision_tree(45))
print("Decision tree result for score 52:", decision_tree(52))
print("Decision tree result for score 85:", decision_tree(85))
````
